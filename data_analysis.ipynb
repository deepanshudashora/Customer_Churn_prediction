```python
sns.countplot(x = 'Gender' , data=df)
```

```python
age18_25 = df.Age[(df.Age <= 30) & (df.Age >= 18)]
age26_35 = df.Age[(df.Age <= 40) & (df.Age >= 30)]
age36_45 = df.Age[(df.Age <= 50) & (df.Age >= 40)]
age46_55 = df.Age[(df.Age <= 60) & (df.Age >= 50)]
age55above = df.Age[df.Age >= 60]

x = ["18-30","30-40","40-50","50-60","60+"]
y = [len(age18_25.values),len(age26_35.values),len(age36_45.values),len(age46_55.values),len(age55above.values)]

sns.barplot(x=x, y=y,data=df)
plt.xlabel("age-groups")
plt.ylabel("number of customer")
plt.title("age  vs  number of customer")
```

```python
sns.barplot(x="Gender", y="CreditScore",data=df)
```

```python
credit1 = df.CreditScore[(df.Age <= 30) & (df.Age >= 18)]
credit2 = df.CreditScore[(df.Age <= 40) & (df.Age >= 30)]
credit3 = df.CreditScore[(df.Age <= 50) & (df.Age >= 40)]
credit4 = df.CreditScore[(df.Age <= 60) & (df.Age >= 50)]
credit5 = df.CreditScore[df.Age >= 60]

x = ["18-30","30-40","40-50","50-60","60+"]
y = [np.average(credit1.values),np.average(credit2.values),np.average(credit3.values),np.average(credit4.values),np.average(credit5.values)]

sns.barplot(x=x, y=y,data=df)
plt.xlabel("age-groups")
plt.ylabel("credit score")
plt.title("age  vs  credit score")
```

```python
b1 = df.Balance[(df.Age <= 30) & (df.Age >= 18)]
b2 = df.Balance[(df.Age <= 40) & (df.Age >= 30)]
b3 = df.Balance[(df.Age <= 50) & (df.Age >= 40)]
b4 = df.Balance[(df.Age <= 60) & (df.Age >= 50)]
b5 = df.Balance[df.Age >= 60]

x = ["18-30","30-40","40-50","50-60","60+"]
y = [np.average(b1.values),np.average(b2.values),np.average(b3.values),np.average(b4.values),np.average(b5.values)]

sns.barplot(x=x, y=y,data=df)
plt.xlabel("age-groups")
plt.ylabel("average balance")
plt.title("age  vs  average balace")
```
